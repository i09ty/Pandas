# Bar Graph

```python
# Bar graph is mainly used to compare between two categories or groups
```

```python
year = [2019,2020,2021,2022]
price =[6700,6800,7000,8000]
plt.bar(year,price, color = 'red')
plt.xlabel('year')
plt.ylabel('price')
plt.title("Year by Year price Compare")
plt.show()
```

```python
langs = ['python','Excel','pandas','matplotlib']
students= [25,18,35,12]
plt.bar(langs,students, color ='blue', width = 0.1)
plt.xlabel('langs')
plt.ylabel('students')
plt.title("Students by Langs Comp.")
plt.show()
```

```python
#plt.title( ) for specifying title of your plot.
#plt.xlabel( ) for labeling x-axis.
#plt.ylabel( ) for labeling y-axis
```

# Line Graph 

```python
#A line chart or line graph is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.
```

```python
p = [1,2,3,4,6]
op = [200,300,500,756,800]
plt.plot(p,op, color ='green', marker ='D', markersize ='10',label = "op")
plt.xlabel("p")
plt.ylabel("op")
plt.title("P VS OP")
plt.show()
```

```python
p = [1,2,3,4,6]
op = [200,300,500,756,800]
plt.plot(p,op,'ro')
plt.xlabel("p")
plt.ylabel("op")
plt.title("P VS OP")
plt.show()
```

#  Histogram 

```python
#Histogram is used to show the frequency distribution of a continuous variable.
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.show()
```
